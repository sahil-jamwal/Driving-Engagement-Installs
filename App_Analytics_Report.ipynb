```python
import pandas as pd
from IPython.display import display

# Replace with the actual path of your Excel file
excel_file_path = "/content/drive/MyDrive/Colab Notebooks/App_analytics_report/App_Analytics_Report.xlsx"

# Load the Excel file into a Pandas ExcelFile object
xls = pd.ExcelFile(excel_file_path)

# Dictionary to store DataFrames with their corresponding table names
dataframes = {}

# Loop through each sheet in the Excel file
for sheet_name in xls.sheet_names:
    # Read the sheet into a DataFrame
    df = pd.read_excel(xls, sheet_name, engine='openpyxl')

    # Check if the DataFrame contains any data
    if not df.empty:
        # Extract the table name from the sheet name
        table_name = sheet_name.replace(" ", "_").lower()

        # Replace '&' with '_' in the table name
        table_name = table_name.replace("&", "_")

        # Store the DataFrame in the dictionary with its table name
        dataframes[table_name] = df

        # Display the table name and DataFrame
        print(f"Table Name: {table_name}")
        display(df)
        print("\n")
```

## Data Cleaning

```python
def analyze_dataframe(df, table_name):
    # Print the table name
    print(f"Table Name: {table_name}")

    # Print column data types and missing values
    for i in df.columns:
        print(f"Column: {i}")
        print(f"Dtype: {df[i].dtypes}")
        print(f"Missing values: {df[i].isna().sum()}")
        print(">>>>>>>>>>>>>>>>")
        print(df[i].value_counts())
        print("________________________")

    # Box plots for numeric columns
    numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
    for col in numeric_columns:
        sns.boxplot(x=col, data=df)
        plt.show()

    # Print a separator line
    print("=====================================================")

# Loop through each DataFrame and call the analyze_dataframe function
for table_name, df in dataframes.items():
    analyze_dataframe(df, table_name)
```
